```
import matplotlib.pyplot as plt           #Fonte: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html
```

```
#      Grafico de Linhas
#Precisamos primeiro sempre passar no mínimo 2 variáveis.
x = [5,8,9,12,15]
y = [0,6,9,11,16]

titulo = "Meu primeiro gráfico com Python" #pode escrever o que quiser no título
plt.title(titulo)

eixoX= "Eixo X"             #pode escrever o que quiser nos eixos
eixoY= "Eixo Y"         #são legendas externas, fora do gráfico
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.plot(x,y)   #aqui pedimos para que construa o grafico
plt.show()      #aqui pedimos para mostrar o grafico construído.
```

```
#       Grafico de Barras
#Precisamos primeiro sempre passar elementos para 2 variáveis.
x = [1,2,3,4,5]   #valores dos gráficos
y = [2,6,9,11,16]   #tamanho das barras no grafico

titulo = "Grafico de Barras"     #pode escrever o que quiser no título
plt.title(titulo)

eixoX= "Eixo X"             #pode escrever o que quiser nos eixos
eixoY= "Eixo Y"
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.bar(x,y)   #aqui pedimos para que construa o grafico
plt.show()      #aqui pedimos para mostrar o grafico construído.
```

```
#       Grafico de Barras - comparação 
#Precisamos primeiro sempre passar elementos para 2 variáveis.
x1 = [1,3,5,7,9]   #valores dos gráficos
y1 = [2,6,9,11,16]   #tamanho das barras no grafico

x2 = [2,4,6,8,10]
y2 = [3,7,10,12,17]
titulo = "Grafico de Barras Comparativo"     #pode escrever o que quiser no título
plt.title(titulo)

eixoX= "Eixo X"             #pode escrever o que quiser nos eixos
eixoY= "Eixo Y"
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.bar(x1,y1, label= "Grupo 1", color = "r")   #aqui pedimos para que construa o grafico, barra em vermelho
plt.bar(x2,y2, label= "Grupo 2", color = "y")    #pedimos que construa o grafico, barra em amarelo
plt.legend()    #pedindo para plotar a legenda 
plt.show()      #aqui pedimos para mostrar o grafico construído.
```

```
#      Grafico dispersão (pontos) Scatterplot
#Precisamos primeiro sempre passar elementos para 2 variáveis.
x = [5,8,9,12,15]
y = [0,6,9,11,16]

titulo = "Grafico Scatterplot"     #pode escrever o que quiser no título
plt.title(titulo)

eixoX= "Eixo X"             #pode escrever o que quiser nos eixos
eixoY= "Eixo Y"
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.scatter(x,y)  #aqui pedimos para que construa o grafico
plt.show()      #aqui pedimos para mostrar o grafico construído.
```

```
#      Grafico dispersão (pontos) Scatterplot  com linhas
#Precisamos primeiro sempre passar elementos para 2 variáveis.
x = [5,8,9,12,15]
y = [0,6,9,11,16]

titulo = "Grafico Scatterplot com linhas"     #pode escrever o que quiser no título
plt.title(titulo)

eixoX= "Eixo X"             #pode escrever o que quiser nos eixos
eixoY= "Eixo Y"
plt.xlabel(eixoX)
plt.ylabel(eixoY)

plt.plot(x,y, color = "k", linestyle= ":")    #insere a linha ligando os pontos, com linha pontilhada, cor preta
plt.scatter(x,y,label = "Meus pontos interligados", color = "r", marker = "*", s=400)  #pontos em vermelho tamanho da estrela 400 e pode ser colocado em variavel
plt.legend()    #plota a legenda interna
plt.show()      #aqui pedimos para mostrar o grafico construído.
plt.savefig("Figura112.png", dpi=300)   #para salvar a figura individualmente pode salvar png ou pdf
```
